```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
for i in company:
    try: 
        companies_name.append(i.find('h2', class_='companyCardWrapper__companyName').text.strip())
    except:
        companies_name.append(np.nan)
    try:
        ratings.append(i.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
    except:
        ratings.append(np.nan)
    try:
        services.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[0])
    except:
        services.append(np.nan)
    try:
        total_employe.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[1])
    except:
        total_employe.append(np.nan)
    try:
        c_type.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[2])
    except:
        c_type.append(np.nan)
    try:
        total_years.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[3])
    except:
        total_years.append(np.nan)
    try:
        location.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[4])
    except:
        location.append(np.nan)
    Highly_rated_for.append(i.find_all('span', class_='companyCardWrapper__ratingValues')[0].text.strip())
    try:
        critacally_rated_for.append(i.find_all('span', class_='companyCardWrapper__ratingValues')[1].text.strip())
    except:
        critacally_rated_for.append(np.nan)
    try:
        reviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[0].text.strip())
    except:
        reviews.append(np.nan)
    try:
        salaries.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
    except:
        salaries.append(np.nan)
    try:
        interviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[2].text.strip())
    except:
        interviews.append(np.nan)
    try:
        jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
    except:
        jobs.append(np.nan)
    try:
        Benefits.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[4].text.strip())
    except:
        Benefits.append(np.nan)
    try:
        Photos.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[5].text.strip())
    except:
        Photos.append(np.nan)
    
d = {"Name": companies_name, 'service_type': services, 'No of Emp': total_employe, 'c_type': c_type, 'total_years': total_years, 'ratings': ratings, 'location': location, 'Highly_rated_for': Highly_rated_for, 'critacally_rated_for': critacally_rated_for, 'reviews': reviews, 'salaries': salaries, 'interviews': interviews, 'jobs': jobs, 'Benefits': Benefits, 'Photos': Photos}

df  = pd.DataFrame(d)
```
